### Exercise #5
- create a sin and logarthmic waveform between 0 and 4 pi then add the two together

```python
x_values = np.linspace(0,2*np.pi,100)

sin = np.sin(x_values)

fig,ax = plt.subplots(1)
ax.plot(x_values,sin,label = "Sine")
ax.set_xlabel("x")
ax.set_title("Sine graph")
ax.legend()
plt.draw()
```

```python
log = np.log(x_values)

fig,ax = plt.subplots(1)
ax.plot(x_values,log, label = "Logarithmic")
ax.set_xlabel("x")
ax.set_title("Logarithmic")
ax.legend()
plt.show()
```

```python
y = np.add(log,sin)

fig,ax = plt.subplots(1)
ax.plot(x_values, y, label = "combined")
ax.set_xlabel("x")
ax.set_title("Combined")
ax.legend()
plt.show()
```

### Exercise #11 - IDk
You will use Numpy and Scipy to perform some analysis on the Iris dataset.

1. Extract the sepal length and sepal width from the Iris dataset and convert them to a Numpy array
2. Use Scipy's interpolation.interp1d to interpolate the sepal width based on the sepal lengths.
3. Plot the values of the sepal width vs. the sepal length with the interpolated function using the provided plotting function.

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from scipy import interpolate

iris = pd.read_csv('https://raw.githubusercontent.com/mwaskom/seaborn-data/master/iris.csv')
print(type(iris))
```

```python
# 1. Extract the sepal length and sepal width from the Iris dataset and convert them to a Numpy array
iris_numpy = iris.to_numpy()
print(iris_numpy)
```

```python
fig, (ax,ax1) = plt.subplots(1,2, figsize = (10,5))
ax.scatter(iris_numpy[:,0],iris_numpy[:,1])
ax.set_xlabel("Sepal Length")
ax.set_title("Scatter Plot")
ax.set_ylabel("Sepal Width")

ax1.plot(iris_numpy[:,0],iris_numpy[:,1])
ax1.set_xlabel("Sepal Length")
ax1.set_title("Scatter Plot")
ax1.set_ylabel("Sepal Width")
plt.show()
```

```python
# 2. Use Scipy's interpolation.interp 1d to interpolate the sepal width based on the sepal lengths
t = np.min(iris_numpy[:,0])
print(t)

t1 = np.max(iris_numpy[:,0])
print(t1)

fig, (ax1,ax2) = plt.subplots(1,2,figsize = (15,5))
ax1.scatter(iris_numpy[:,0],iris_numpy[:,1])
plt.suptitle("Sepal Length vs. Sepal Width")

ax1.set_xlabel("Sepal Length")
ax1.set_ylabel("Sepal Width")
ax1.set_title("Scatter Plot")

f = interpolate.interp1d(iris_numpy[:,0],iris_numpy[:,1], kind = 'nearest')
x = np.linspace(4.3,7.9,100)
y = f(x)

ax2.plot(x,y)
ax2.set_xlabel("Sepal Length")
ax2.set_ylabel("Sepal Width")
ax2.set_title("Interpolated Function of the Data")
```

```python
# 3. Plot the Values of the sepal witdh vs. length with the interpolated function using the provided functions
fig, ax = plt.subplots(1)
ax.scatter(iris_numpy[:,0],iris_numpy[:,1])
ax.plot(x,y, 'g')
plt.show()
```

```python
# Another way to interpolating 
def function(x,y):
    y_values = interpolate.interp1d(x, y)
    x_values = np.linspace(np.min(x),np.max(x), 100)
    f = y_values(x_values)


    fig, ax = plt.subplots(1, figsize = (10,5))
    ax.scatter(x,y)
    ax.plot(x_values, f, color = 'g')
    ax.set_xlabel("Sepal Length")
    ax.set_ylabel("Sepal Width")
    ax.set_title("Interpolated Function of the Data")
    plt.show()

function(iris_numpy[:,0], iris_numpy[:,1])
```
